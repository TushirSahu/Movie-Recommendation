```python
indices=pd.DataFrame(df1).drop_duplicates()
# df2.columns
indices['tags']=df2['overview']
indices.drop(['cast','crew'],axis=1)
indices
```

```python
tf=TfidfVectorizer(stop_words='english')
indices['tags']=indices['tags'].fillna('')
tf_vector=tf.fit_transform(indices['tags'])
```

```python
tf_vector.shape
```

```python
df1
```

```python
cosine_sim=linear_kernel(tf_vector,tf_vector)
```

```python
def get_recommendations(movie):
    idx=indices[indices['title']==movie].index[0]
    sim_scores=sorted(list(enumerate(cosine_sim[idx])),key=lambda x:x[1],reverse=True)
    for i in sim_scores[1:7]:
        print(df2.iloc[i[0]].title)
        # print(indices.iloc[i[0]].id)
#     return df2['title'].iloc[movie_indices]
```

```python
get_recommendations('The Avengers')
```
